# Part II: Object-Oriented Programming

## 5. Classes & Objects

### Creating a Class

```python
# Simple class
class Dog:
    def __init__(self, name):
        self.name = name  # Attribute
    
    def bark(self):
        return f'{self.name} says Woof!'

# Create object
my_dog = Dog('Buddy')
print(my_dog.bark())
```

```python
# Oops in python
```

```python
# Class with multiple attributes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        return f'I am {self.name}, {self.age} years old'

alice = Person('Alice', 25)
print(alice.introduce())
```

### Class Methods

```python
# Instance method (uses self)
class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, n):
        self.result += n
        return self.result

calc = Calculator()
print(calc.add(5))  # 5
print(calc.add(3))  # 8
```

### Inheritance

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return 'Some sound'

# Child class
class Cat(Animal):
    def speak(self):
        return 'Meow!'

cat = Cat('Whiskers')
print(f'{cat.name} says: {cat.speak()}')
```

# Part III: Data Structures & Algorithms

## 10. Time Complexity (Big-O)

### What is Big-O?

Measures how runtime grows with input size.

- **O(1)** - Constant (best)
- **O(log n)** - Logarithmic
- **O(n)** - Linear
- **O(n²)** - Quadratic (slow)

```python
# O(1) - Constant time
def get_first(lst):
    return lst[0]  # Always same time

print(get_first([1, 2, 3, 4, 5]))
```

```python
# O(n) - Linear time
def find_in_list(lst, target):
    for item in lst:  # Check each item
        if item == target:
            return True
    return False

print(find_in_list([1, 2, 3, 4, 5], 3))
```

```python
# O(n²) - Quadratic time (nested loops)
def print_pairs(lst):
    for i in lst:        # Loop 1
        for j in lst:    # Loop 2 (inside loop 1)
            print(f'({i}, {j})')

print_pairs([1, 2, 3])
```

## 11. Lists (Dynamic Arrays)

### List Operations & Speed

| Operation | Time | Notes |
|-----------|------|-------|
| `lst[i]` | O(1) | Fast |
| `lst.append(x)` | O(1) | Fast |
| `lst.insert(0, x)` | O(n) | Slow |
| `x in lst` | O(n) | Slow |

```python
# Fast: append to end
numbers = []
for i in range(5):
    numbers.append(i)  # O(1) each
print(numbers)
```

```python
# Slow: insert at beginning
numbers = [1, 2, 3]
numbers.insert(0, 0)  # O(n) - shifts all
print(numbers)
```

### List Comprehension

```python
# Traditional way
squares = []
for x in range(5):
    squares.append(x**2)
print(squares)
```

```python
# List comprehension (better!)
squares = [x**2 for x in range(5)]
print(squares)
```

```python
# With condition
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

## 12. Stacks & Queues

### Stack (LIFO - Last In, First Out)

Like a stack of plates - add/remove from top only.

```python
# Stack using list
stack = []

# Push (add to top)
stack.append(1)
stack.append(2)
stack.append(3)
print(f'Stack: {stack}')
```

```python
# Pop (remove from top)
top = stack.pop()
print(f'Popped: {top}')
print(f'Stack now: {stack}')
```

### Queue (FIFO - First In, First Out)

Like a line - first in is first out.

```python
# Queue using deque (efficient!)
from collections import deque

queue = deque()
queue.append(1)  # Add to back
queue.append(2)
queue.append(3)
print(f'Queue: {list(queue)}')
```

```python
# Dequeue (remove from front)
first = queue.popleft()
print(f'Removed: {first}')
print(f'Queue now: {list(queue)}')
```

## 14. Trees

### Binary Tree

Each node has at most 2 children (left & right).

```python
# Tree node
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
# Build simple tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)

print(f'Root: {root.value}')
print(f'Left: {root.left.value}')
print(f'Right: {root.right.value}')
```

### Tree Traversal (Inorder)

```python
# Inorder: Left -> Root -> Right
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.value, end=' ')
    inorder(node.right)

inorder(root)
```

## 15. Sorting Algorithms

### Bubble Sort (O(n²))

```python
# Bubble sort - simple but slow
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort([64, 34, 25, 12]))
```

### Python's Built-in Sort (O(n log n))

```python
# Use Python's sort (fastest!)
arr = [64, 34, 25, 12, 22]

# Method 1: sorted() - returns new list
sorted_arr = sorted(arr)
print(f'Sorted: {sorted_arr}')
print(f'Original: {arr}')
```

```python
# Method 2: sort() - modifies in place
arr = [64, 34, 25, 12, 22]
arr.sort()
print(f'After sort(): {arr}')
```

### Binary Search (O(log n))

```python
# Binary search - MUST be sorted!
def binary_search(arr, target):
    left, right = 0, len(arr)-1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1, 3, 5, 7, 9]
print(f'Find 5: {binary_search(arr, 5)}')
```
