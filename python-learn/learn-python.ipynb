### Dictionary

````python

student = {
    "name": {"Himanshu","Tanishq"},
    "age":{20,21},
    "grade": {"A","B"}
}

print(student)
print(type(student))
````

````python
# Add new value this dictionary

student["city"] = {"Delhi","Patna"}

print(student)

````

````python
# Get all keys
print(student.keys())       # Output: dict_keys(['name', 'age', 'grade'])

# Get all values
print(student.values())     # Output: dict_values(['Himanshu', 20, 'A'])

# Get key-value pairs
print(student.items())      # Output: dict_items([('name', 'Himanshu'), ('age', 20), ('grade', 'A')])

# Safe access (returns None if key doesn't exist)
print(student.get("phone"))  # Output: None
````

````python
import numpy as np

arr_values = list(student.values())
arr_values = np.array(arr_values)
print("data type is:",type(arr_values))
print(arr_values)
````

### In python we do negative indexing
```
Array:        [10,  20,  30,  40,  50]
Positive:      0    1    2    3    4
Negative:     -5   -4   -3   -2   -1
```

````python
arr = np.array([10,20,30,40,50])
print("first index using positive indexing :", arr[0])
print("first index using negative indexing :", arr[-5])
print("last index using positive indexing :", arr[4])
print("last index using negative indexing :", arr[-1])
````
